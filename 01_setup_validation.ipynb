```python
# Célula 2 — Verificação de ambiente (vai para o artigo como Tabela I)
import torch, platform, transformers, nemo

config = {
    "Python": platform.python_version(),
    "PyTorch": torch.__version__,
    "CUDA disponível": torch.cuda.is_available(),
    "GPU": torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU",
    "Transformers": transformers.__version__,
    "NeMo": nemo.__version__,
}

for k, v in config.items():
    print(f"{k:25s}: {v}")
```

```python
# Célula 3 — Carregamento dos modelos com tempo de inicialização
import time
from transformers import AutoProcessor, SeamlessM4Tv2Model

print("Carregando SeamlessM4T...")
t0 = time.time()
processor_seamless = AutoProcessor.from_pretrained("facebook/seamless-m4t-v2-large")
model_seamless = SeamlessM4Tv2Model.from_pretrained("facebook/seamless-m4t-v2-large")
print(f"SeamlessM4T carregado em {time.time()-t0:.1f}s")
```

```python
# Célula 4 — NeMo Canary
import nemo.collections.asr as nemo_asr

print("Carregando NeMo Canary...")
t0 = time.time()
model_canary = nemo_asr.models.EncDecMultiTaskModel.from_pretrained("nvidia/canary-1b")
print(f"NeMo Canary carregado em {time.time()-t0:.1f}s")
```

```python
# Célula 5 — Parâmetros dos modelos (Tabela I do artigo)
import pandas as pd

def contar_parametros(modelo):
    return sum(p.numel() for p in modelo.parameters()) / 1e6

tabela_modelos = pd.DataFrame([
    {"Modelo": "SeamlessM4T-v2-Large", "Parâmetros (M)": contar_parametros(model_seamless), "Tarefa": "S2TT / ASR"},
    {"Modelo": "NeMo Canary-1B",       "Parâmetros (M)": 1000,                               "Tarefa": "ASR / S2TT"},
])
print(tabela_modelos.to_string(index=False))
# → Esta tabela entra literalmente como Table I no artigo
```
